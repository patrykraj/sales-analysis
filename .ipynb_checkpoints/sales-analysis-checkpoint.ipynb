```python
df['Revenue'] = df['Quantity'] * df['UnitPrice']
```

```python
df
```

```python
df_clean = df[(df['Quantity'] > 0) & (df['UnitPrice'] > 0) & (df['CustomerID'].notna())].copy()
```

```python
total_revenue = df_clean['Revenue'].sum()
total_revenue
```

```python
top_products = df_clean.groupby('Description')['Revenue'].sum().sort_values(ascending=False).head(10)
top_products
```

```python
df_clean['InvoiceDate'] = pd.to_datetime(df_clean['InvoiceDate'])
df_clean['CustomerID'] = df_clean['CustomerID'].astype(int)
df_clean['Country'] = df_clean['Country'].astype('category')
```

```python
top_customers = df_clean.groupby('CustomerID')['Revenue'].sum().sort_values(ascending=False).head(10)
top_customers
```

```python
df_clean.loc[:, 'InvoiceDate'] = pd.to_datetime(df_clean['InvoiceDate'])
monthly_revenue = df_clean.resample('ME', on='InvoiceDate')['Revenue'].sum()
monthly_revenue.plot(kind='line')
```

```python
revenue_by_country = df_clean.groupby('Country')['Revenue'].sum().sort_values(ascending=False)
revenue_by_country
```
